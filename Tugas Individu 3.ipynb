## Nomor 1

```python
#Import library
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Memasukan data
X = np.array([[25, 3000, 12], [34,4500, 24], [29, 5000, 26],
              [45, 6000, 6], [32, 5500, 18], [50, 7000, 48],
              [40, 4000, 8], [28, 3200, 14], [35, 4800, 30],
              [22, 2500, 10]])
y = np.array([[0,1,0,1,0,0,1,0,1,0]])
```

```python
y = y.reshape(-1)  # atau y = y.flatten()
```

```python
#import fungsi SVM dari library sklearn
#menggunakan kernel linear karena menggunakan persamaan linear
model = SVC(kernel='linear')
```

```python
#Latih model menggunakan dataset yang sudah disiapkan
#gunakan data x sebagai fitur masukan dan data y sebagai keluran
model.fit(X,y)
```

```python
#lihat support vectors pad model
print(f'Support vectors:\n {model.support_vectors_}\n')
print(f'Support vektor pada masing-masing kelas:\n {model.n_support_}')
```

```python
#Menampilkan data dan hyperplane svm

sns.scatterplot(x=X[:,0], y=X[:,1], hue=y, s=30)
w = model.coef_[0]
b = model.intercept_[0]

#buat skala pada sumbu x dari 0-10
x_points = np.linspace(1,50)

#buat nilai pada sumbu y menggunakan nilai x
y_points = -(w[0]/w[1]) * x_points - b/w[1]

#plot hyperplane menggunakan warna merah
plt.plot(x_points, y_points, c ='b')
plt.show()
```

```python
#Lakukan prediksi menggunakan data baru: Umur= 35, Pendapatan Bulanan=4700, dan Lama Berlangganan 23
X_uji = np.array([[35, 4700, 23]])
```

```python
prediksi = model.predict(X_uji)
print(f'Hasil prediksi SVM: {prediksi}')
```

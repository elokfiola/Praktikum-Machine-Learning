## Soal 1
Lakukan klaseterisasi dengan Hierarchical Agglomerative Clustering dengan metric Euclidean pada dataset sebagai berikut : (2, 2), (3, 3), (4, 2), (8, 7), (9, 8), (10, 7), (15, 2), (16, 3), (17, 2). Gunakan metode  pengelompokkan Single Linkage, Complete dan Average Linkage dan tampilkan dendrogram ! (ikuti langkah-langkah pada D1).


```python
#Import library
import numpy as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#Import Dataset
x = [2,3,4,8,9,10,15,16,17]
y = [2,3,2,7,8,7,2,3,2]

plt.scatter(x, y)
plt.show()
```

```python
#Pengemlopokan aglomeratif Single Linkage
data = list(zip(x,y))

linkage_data_s = linkage(data, method='single', metric='euclidean')
dendrogram(linkage_data_s)
```

```python
#Pengemlopokan aglomeratif complete linkage
data = list(zip(x,y))

linkage_data_c = linkage(data, method='complete', metric='euclidean')
dendrogram(linkage_data_c)
```

```python
#Pengemlopokan aglomeratif
data = list(zip(x,y))

linkage_data_a = linkage(data, method='average', metric='euclidean')
dendrogram(linkage_data_a)
```

## Soal 2
Dengan dataset yang sama pada nomor 1 lakukan klasterisasi menggunakan Gunakan metode  pengelompokkan Single Linkage, Complete dan Average Linkage dan tampilkan hasil berdasarkan label dengan 3 klaster ! (ikuti langkah-langkah pada D2).


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
#Import Dataset
x = [2,3,4,8,9,10,15,16,17]
y = [2,3,2,7,8,7,2,3,2]

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [2,3,4,8,9,10,15,16,17]
y = [2,3,2,7,8,7,2,3,2]

# Membuat scatter plot
plt.scatter(x, y)

# Menambahkan label untuk setiap titik
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0, 5), ha='center')

# Menampilkan plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Annotations')
plt.grid()
plt.show()
```

```python
#Pengelompokkan aglomeratif single linkage
data = list(zip(x,y))

hierarchical_cluster_s = AgglomerativeClustering(n_clusters=3, linkage='single')
labels_s = hierarchical_cluster_s.fit_predict(data)
```

```python
#Tampilan hasil pengelompokan dari label
print(labels_s)

plt.scatter(x, y, c = labels_s)
plt.show()
```

```python
#Pengelompokkan aglomeratif complete linkage
data = list(zip(x,y))

hierarchical_cluster_c = AgglomerativeClustering(n_clusters=3, linkage='complete')
labels_c = hierarchical_cluster_c.fit_predict(data)
```

```python
#Tampilan hasil pengelompokan dari label
print(labels_c)

plt.scatter(x, y, c = labels_c)
plt.show()
```

```python
#Pengelompokkan aglomeratif average linkage
data = list(zip(x,y))

hierarchical_cluster_a = AgglomerativeClustering(n_clusters=3, linkage='average')
labels_a = hierarchical_cluster_a.fit_predict(data)
```

```python
#Tampilan hasil pengelompokan dari label
print(labels_a)

plt.scatter(x, y, c = labels_a)
plt.show()
```

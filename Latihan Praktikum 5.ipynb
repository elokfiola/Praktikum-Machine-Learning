## Implementasi Hierarchical Agglomerative Clustering

```python
#Import library
import numpy as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
#Import Dataset
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [4, 5, 10, 4, 3, 11, 14, 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

# Membuat scatter plot
plt.scatter(x, y)

# Menambahkan label untuk setiap titik
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0, 5), ha='center')

# Menampilkan plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Annotations')
plt.grid()
plt.show()
```

```python
#Pengemlopokan aglomeratif
data = list(zip(x,y))

linkage_data = linkage(data, method='complete', metric='euclidean')
dendrogram(linkage_data)
```

## Implementasi algoritma clustering aglomeratif dengan AgglomerativeClustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
```

```python
#Import Dataset
x = [4,5,10,4,3,11,14,6,10,12]
y = [21,19,24,17,16,25,24,22,21,21]

plt.scatter(x, y)
plt.show()
```

```python
#Pengelompokkan aglomeratif
data = list(zip(x,y))

hierarchical_cluster = AgglomerativeClustering(n_clusters=2, linkage='complete')
labels = hierarchical_cluster.fit_predict(data)
```

```python
#Tampilan hasil pengelompokan dari label
print(labels)

plt.scatter(x, y, c = labels)
plt.show()
```

```python
#Melakukan klustrisaksi pengimplementsian AC menggunakan mahanntam
#Pengelompokkan manhattan
data = list(zip(x,y))

hierarchical_cluster = AgglomerativeClustering(n_clusters=2, linkage='complete', metric='manhattan')
labels = hierarchical_cluster.fit_predict(data)

#Tampilan hasil pengelompokan dari label
print(labels)

plt.scatter(x, y, c = labels)
plt.show()
```

```python
#Melakukan klustrisaksi pengimplementsian AC menggunakan mahanntam
#Pengelompokkan manhattan
data = list(zip(x,y))

# Ubah linkage menjadi 'average' agar affinity='manhattan' dapat digunakan.
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, linkage='average', metric='manhattan')
labels = hierarchical_cluster.fit_predict(data)

#Tampilan hasil pengelompokan dari label
print(labels)

plt.scatter(x, y, c = labels)
plt.show()
```
